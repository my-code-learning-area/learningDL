```python
import matplotlib.pyplot as mat
import keras
```

```python
import pandas as pd
train = pd.read_csv('./LP-IV-datasets/MNIST/mnist_train.csv')
test = pd.read_csv('./LP-IV-datasets/MNIST/mnist_test.csv')
train
```

```python
x_train = train.drop('label', axis=1)
x_test = test.drop('label', axis=1)

x_train
```

```python
y_train = train['label']
y_test = test['label']

y_train
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
# convert y to categorical
y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)
```

```python
# what this do?
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
train_shape_y = x_train.shape[1]
```

```python
?model.fit
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import SGD
```

```python
model = Sequential([
    Dense(128, input_shape=(train_shape_y,), activation="relu"),
    Dense(64, activation="relu"),
    Dropout(0.2),
    Dense(10, activation="softmax")
])

model.compile(loss="categorical_crossentropy", optimizer=SGD(learning_rate=0.1, momentum=0), metrics=['accuracy'])
model.summary()
```

```python
history=model.fit(x=x_train, y=y_train, batch_size=512, epochs=11)
```

```python
acc, los = model.evaluate(x_test, y_test)
print(acc)
print(los)
```

```python
mat.plot(history.history['accuracy'])
```

```python
mat.plot(history.history['loss'])
```
